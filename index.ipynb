```python
def clean(col_name):
    cleaned = col_name.strip()
    return cleaned
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
df.DATE = pd.to_datetime(df.DATE)
df['Dayofweek'] = df.DATE.dt.dayofweek
grouped = df.groupby('Dayofweek').sum()
print(grouped)
grouped.plot(kind='bar')
#Your code here
#index by weekday
#drop columns
#Num_Lines

#plt.hist(x, bins = 10)

#plt.xlabel('Weekday')
#plt.ylabel('Frequency of Values')
#plt.title('Subway Traffic by Day of the Week')
#plt.show()
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
grouped = grouped.reset_index()
print(grouped.head(8)) #Your code here
grouped['Weekendvs.Weekday'] = grouped.Dayofweek.map({0:False,1:False,2:False,3:False,4:False,5:True,6:True})
wkendvwkday = grouped.groupby('Weekendvs.Weekday').mean()
print(wkendvwkday.head(8))
wkendvwkday.plot(kind='bar')
```
